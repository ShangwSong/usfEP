## A code for calculating euler vecotors for plate motion from GPS velocity

```python
import numpy as np 
from usfEP import euler_gps,euler_pole,euler_calculate
%pylab inline
```

usfEP has the following modules:

    *euler_pole : class for holding euler vectors
    
    *euler_calculate : geodetic methods for calculating neccesary vectors
    
    *euler_gps: class for holding gps data and inversion methods
 
First we need to initialize a network object, we will use pacific_2.txt in the tutorialdata directory. 

A function for importing text files in the format of pacific_2.txt has been provided, but getting data into network structure should be fairly straightforward. 

```python
net = euler_gps.txt2network('tutorial_data/pacific_2.txt','pacific')
```

A network is a collection of stations:

A station has the following attributes:

*name

*position (lon,lat)

*velocity (east,north,up)

*covariance

We can plot the stations like so:

```python
net.plotmap()
```

So now we can see the velocity of the stations

We will inver the GPS vectors for the euler vector using weighted least squares from the pjInvert package. 


```python
pole = net.invert()
```

We can plot this pole on a map

```python
pole.plot()
```
